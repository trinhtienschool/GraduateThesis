```python
from google.colab import drive
import ast
import os
from glove import Corpus, Glove
```

```python
#Load corpus
#Corpus is a 2d array

def load_corpus_file():
  f = open('preprocessed_corpus.txt', "r")
  corpus = []
  for doc in f:
      corpus.append(ast.literal_eval(doc))
  f.close()
  return corpus

corpus_input_data  = load_corpus_file()
```

```python
#Save pre-trained GloVe
def save_glove_embedding(file_name):
  f = open(file_name, "w")
  for word, index in corpus.dictionary.items():
    f.write(word + ' ')
    value = ''
    for j in range(len(glove.word_vectors[index])):
      value += ' ' + str(glove.word_vectors[index,j])
    value = value[1:len(value)].strip()  
    f.write(value + '\n')    
  f.close()
  print("Saved!")
```

**Train GloVe embedding**

```python
#Make co-occurrent matrix
corpus = Corpus() 
corpus.fit(corpus_input_data, window=10)
```

Thực hiện Glove Embedding

```python
#Train GloVe embedding
glove = Glove(no_components=300, learning_rate=0.05) 
glove.fit(corpus.matrix, epochs=100, verbose=True)
glove.add_dictionary(corpus.dictionary)
save_glove_embedding("Glove_pre_trained_300d.txt")
```

**Test similar**

```python
#Check most similar
glove.most_similar('phone',number=10)
```
